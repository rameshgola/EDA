DATA PREPERATION

```python
#converting some binery veriables yes or no to 1 or 0
varlist = ['PhoneService', 'PaperlessBilling', 'Churn', 'Partner', 'Dependents']
    
def binery_map(x):
    return x.map({'Yes':1,'No':0})
df[varlist] = df[varlist].apply(binary_map)
```

```python
df.head()
```

```python
df[['Contract','PaymentMethod']].head()
```

```python
#creating the categorical values as dummy
dummy = pd.get_dummies(df[['Contract','PaymentMethod','gender','InternetService']],drop_first=True)
df = pd.concat([df,dummy],axis=1)
```
